```python
# First, import the needed packages
import pandas as pd
import bar_chart_race as bcr
```

```python
premier = pd.read_csv('premier.csv', parse_dates=True, index_col='date')
```

We need to transpose the input, bar-chart-race takes the dates as rows and the cumulative data as columns.

```python
premier_transposed = premier.transpose()
```

```python
html = bcr.bar_chart_race(premier_transposed, filename=None)
```
